# Association Rule Mining

Apa itu Association Rule Mining

Association Rule Mining merupakan salah satu teknik yang penting dan diggunakan dalam mencari pola dalam database. Tujuan yang ditmakan dalam Association Rule Mining adalah mengetahui hubungan korelasi di antara beberapa item yang berbeda dari database. Dengan Association rule mining kita dapat menentukan pola dan banyak kombinasi dalam database tersebut. Setelah itu kita akan dapat mengetahui banyak kombinasi dalam waktu yang sama. Teknik ini sering dilakukan dalam database ukuran besar dan dapat berlaku di berbagai bidang dalam kehidupan

Tipe dari Association Rule Mining

Menurut(GeeksforGeeks,2022) Terdapat 4 jenis Association Rule Mining yaitu

1. Multi-relational association rules
2. Generalized association rules
3. Quantitative association rules
4. Interval information association rules

Pengaplikasian dari Association Rule Mining

Market Basket Analysis

Menurut (Algoritma 2022) Market basket analysis adalah salah satu dari association rule mining yang digunakan untuk mengungkap pola pembelian dalam bidang ritel apa pun. Tujuannya adalah memahami perilaku konsumen dengan mengidentifikasi hubungan antara barang-barang yang dibeli. Contohnya, konsumen yang membeli roti juga akan cenderung membeli selai. Jadi, teknik ini secara kuantitatif akan menetapkan bahwa ada hubungan antara roti dan selai. Dengan adanya informasi ini pemilik retail dapat mengubah posisi toko yang optimal

Cross Marketing

Cross marketing adalah tindakan melakukan kerjasama dengan bisnis lain untuk mengembangkan bisnis yang dimiliki. Contoh bisnis selai melakukan kerjasama dengan bisnis roti untuk mengembangkan daya jual. Biasanya perusahaan yang melakukan cross marketing adalah product yang saling berhubungan

Algoritma dari Association Rule Mining

Alpriori Algoritm

Algoritma Alpriori adalah algoritma yang diggunakan untuk menghitungaturan asosiasi antar objek. Aturan asosiasi menjelaskan bagaimana dua atau lebih objek terkait satu sama lain.

Di atas menunjukan bahwa kombinasi product dari dua belah pihak dapat mempengaruhi pasanganya dalam penjualan

Eclat Algorythm

Algoritma ECLAT adalah singkatan dari Equivalence Class Clustering dan bottom-up Lattice Traversal. Ini adalah salah satu metode penambangan Aturan Asosiasi yang populer. Ini adalah versi algoritma Apriori yang lebih efisien dan terukur. Sementara algoritme Apriori bekerja dalam arti horizontal dengan meniru Pencarian Luas-Pertama dari sebuah grafik, algoritme ECLAT bekerja secara vertikal seperti Pencarian Kedalaman-Pertama dari sebuah grafik. Pendekatan vertikal algoritme ECLAT ini menjadikannya algoritme yang lebih cepat daripada algoritme Apriori.

Kelebihan dibanding Alpriory
1. Persyaratan Memori: Karena algoritme ECLAT menggunakan pendekatan Pencarian Kedalaman-Pertama, algoritme ini menggunakan lebih sedikit memori daripada algoritme Apriori.
2. Kecepatan: Algoritma ECLAT biasanya lebih cepat daripada algoritma Apriori.
3. Jumlah Komputasi: Algoritme ECLAT tidak melibatkan pemindaian data berulang kali untuk menghitung nilai dukungan individu.

FP Growth Algorithm

Algoritma ini merupakan penyempurnaan dari metode Apriori. Model sering dibuat tanpa harus membuat kandidat. mewakili database sebagai pohon yang dikenal sebagai pohon sampel biasa atau pohon FP. Struktur pohon ini akan menjaga hubungan antar koleksi item. Basis data terfragmentasi menggunakan elemen yang sering digunakan. Fragmen ini disebut "fragmen sampel". Kumpulan yang sering dari model fragmentasi ini dianalisis. Sehingga dengan metode ini pencarian frequent itemset relatif berkurang.
